```python
from pandas import json_normalize
import json

#return json object
filename = 'leetcodeProblems.json'
obj = json.load(open(filename))
problem_list = obj['stat_status_pairs']
problem_list[0]
```

```python
df = json_normalize(problem_list)
df.head()
```

```python
df.to_csv('leetcodeProblems.csv', index=False)
```

```python
free = df[df['paid_only'] == False]
free.head()

free.to_csv('free_leetcodeProblems.csv', index=False)
```

```python
easy = free[free['difficulty.level'] == 1]
medium = free[free['difficulty.level'] == 2]
hard = free[free['difficulty.level'] == 3]

columns_to_keep = [
    'stat.question_id',
    'stat.frontend_question_id',
    'stat.question__title',
    'stat.question__title_slug',
    'stat.total_acs',
    'stat.total_submitted',
]

easy = easy[columns_to_keep]
medium = medium[columns_to_keep]
hard = hard[columns_to_keep]
```

```python
easy.to_csv('easy_free_leetcodeProblems.csv', index=False)
medium.to_csv('medium_free_leetcodeProblems.csv', index=False)
hard.to_csv('hard_free_leetcodeProblems.csv', index=False)
```

```python
def open_random_problem(df):
    import webbrowser
    problem = df.sample(1)
    slug = problem['stat.question__title_slug'].values[0]
    url = f'https://leetcode.com/problems/{slug}/'
    webbrowser.open(url)
```

```python
open_random_problem(easy)
```
